```python
from sklearn.datasets import load_iris

df = load_iris(as_frame=True)  
df = df.frame   # Convert to pandas DataFrame
```

```python
# selection with loc and iloc

print(df.loc[0, 'sepal length (cm)'])    # first row, specific column

print(df.iloc[0, 0])                     # same as above
```

```python
# Selecting multiple rows/columns
df.loc[0:4, ['sepal length (cm)', 'target']]   #unlike Python slicing with lists, loc includes the end value
```

```python
# FILTERING
# Filter rows where sepal length > 7
df[df['sepal length (cm)'] > 7]
```

```python
# Filter rows of species 1
df[df['target'] == 1].head()
```

```python
# AGGREGATION
# Mean sepal length per species
df.groupby('target')['sepal length (cm)'].mean()
```

```python
# SORTING
# Sort by sepal length
df.sort_values(by='sepal length (cm)', ascending=False).head(10)
```

```python
# Calculate sepal area (length × width)
df['sepal area'] = df['sepal length (cm)'] * df['sepal width (cm)']
df.head()
```

```python
# Write DataFrame to CSV
df.to_csv("iris.csv", index=False)
```
