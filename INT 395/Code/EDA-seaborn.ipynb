```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the built-in Titanic dataset from seaborn
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.histplot(data=titanic, x='age', bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
# count of passengers by class
sns.countplot(x='class', data=titanic, hue='class', palette='pastel')
plt.title('Passenger Count by Class')
plt.show()
```

```python
sns.countplot(data=titanic, x='sex', hue='survived', palette='Set2')
plt.title('Survival Count by Gender')
plt.show()
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived', alpha=0.7)
plt.title('Age vs Fare (colored by Survival)')
plt.show()
```

```python
sns.boxplot(data=titanic, y='age')
plt.title('Distribution of Passenger Ages')
plt.show()
```

```python
sns.boxplot(data=titanic, x='fare')
plt.title('Distribution of Fare (Horizontal)')
plt.show()
```

```python
#catplot is used for categorical comparisons across multiple dimensions.
sns.catplot(data=titanic, x='class', hue='sex', col='survived', kind='count', height=4, aspect=0.9)
```

```python
# Select numeric columns only
numeric_data = titanic.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr = numeric_data.corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='Blues', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(data=titanic, vars=['age', 'fare', 'pclass'], hue='survived')
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```
