# HANDLING OUTLIERS

```python
# Select only numeric columns for outlier analysis
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
print(num_cols)
```

```python
# Plot boxplots for all numeric features
plt.figure(figsize=(12, 6))
df[num_cols].boxplot()
plt.title('Boxplot for Numeric Columns')
plt.show()
```

```python
#Return lower and upper IQR limits for detecting outliers
def get_outlier_limits(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return lower, upper
```

```python
# Remove Outliers
def remove_outliers(data, column):
    lower, upper = get_outlier_limits(data, column)
    filtered = data[(data[column] >= lower) & (data[column] <= upper)]
    return filtered

df_removed = remove_outliers(df.copy(), 'fare')
print(f"Removed {df.shape[0] - df_removed.shape[0]} outlier rows.")
df_removed
```

```python
# Cap Outliers (Winsorization) -- Replaces outliers with boundary values
# np.where(condition, value_if_true, value_if_false)
def cap_outliers(data, column):
    lower, upper = get_outlier_limits(data, column)
    data[column] = np.where(data[column] < lower, lower, np.where(data[column] > upper, upper, data[column]))
    return data

df_capped = cap_outliers(df.copy(), 'fare')
print("Outliers capped to boundary values.")
df_capped
```

```python
# Log Transformation (for skewed data)
#Applies log(1 + x) transformation to reduce effect of outliers
def transform_outliers_log(data, column):
    data[column + '_log'] = np.log1p(data[column])  
    return data

df_log = transform_outliers_log(df.copy(), 'fare')
print("Log transformation applied. New column:", 'fare_log')
df_log
```
