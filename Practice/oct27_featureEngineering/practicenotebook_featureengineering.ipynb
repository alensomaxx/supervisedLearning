```python
import pandas as pd
import numpy as np
df= pd.read_csv("HousePrices.csv")
# Example: Creating a new feature 'total_rooms' by adding bedrooms and bathrooms
df['total_rooms'] = df['bedrooms'] + df['bathrooms']
df.head()
```

```python
# Example: Creating a new feature 'house_age' by subtracting the year built from the current year
current_year = pd.to_datetime('now').year
df['house_age'] = current_year - df['yr_built']
print(df[['yr_built', 'house_age']].head())
```

```python
# Example: Creating a new feature 'price_per_sqft' by dividing price by sqft_living
df['price_per_sqft'] = df['price'] / df['sqft_living']
print(df[['price', 'sqft_living', 'price_per_sqft']].head())
```

```python
# Logarithmic transformation of the 'price' column
df['log_price'] = np.log(df['price'])
df.head()
```

```python
# Logarithmic transformation of the 'price' column
df['log_sqftlog'] = np.log(df['sqft_lot'])
df.head()
```

```python
# Square root transforming the 'price' variable
df['SquareRoot_price'] = np.sqrt(df['price'])

# Displaying the DataFrame with the new feature
print("DataFrame with square root transformed 'price':")
print(df[['price', 'SquareRoot_price']])
```

```python
from scipy.stats import boxcox

# Applying Box-Cox transformation to 'sales' variable
df['BoxCox_Price'], _ = boxcox(df['sqft_living'])

# Displaying the DataFrame with the Box-Cox transformed 'sales' variable
print("DataFrame with box-cox transformed price:")
print(df[['sqft_living', 'BoxCox_Price']])
```

```python
from sklearn.preprocessing import MinMaxScaler

# Scaling numeric features using min-max scaling
scaler = MinMaxScaler()
df[['sqft_living', 'sqft_lot']] = scaler.fit_transform(df[['sqft_living', 'sqft_lot']])
print(df)
```

```python
# Sample DataFrame
data = {'color': ['red', 'blue', 'green', 'red', 'green']}##dictionary (data) key value pair, only 1 key and 1 value list
df2 = pd.DataFrame(data)##dataframe is created: df2

# Before one-hot encoding
print("Original DataFrame:")
print(df2)

# Apply one-hot encoding
df2_encoded = pd.get_dummies(df2, columns=['color'], prefix='color')## encoded output is stored df2_encode, get_dummies method for ohe on the column, color

# After one-hot encoding
print("\nDataFrame after one-hot encoding:")
print(df2_encoded)
```

```python
from sklearn.preprocessing import LabelEncoder

# Sample DataFrame
data = {'size': ['small', 'medium', 'large', 'medium', 'small']}##same
df1 = pd.DataFrame(data)

# Before label encoding
print("Original DataFrame:")
print(df1)

# Apply label encoding
label_encoder = LabelEncoder()##initiated
df1['size_encoded'] = label_encoder.fit_transform(df1['size'])

# After label encoding
print("\nDataFrame after label encoding:")
print(df1)
```

```python
from sklearn.preprocessing import LabelEncoder

# Sample DataFrame
data = {'Floor' : ['first', 'second','third','forth']}##same
df3 = pd.DataFrame(data)

# Before label encoding
print("Original DataFrame:")
print(df3)

# Apply label encoding
label_encoder = LabelEncoder()##initiated
df3['size_encoded'] = label_encoder.fit_transform(df3['Floor'])

# After label encoding
print("\nDataFrame after label encoding:")
print(df3)
```
