roc curve method, calculates multiple POSSIBLE threshold values, for each threshold value true positive +ve and false +ve rate is calculated.

(in the next part) an optimum threshold value is calculated. This value is that threshold value, at which the true +ve rate is highest, and the false +ve rate is lowest. At this value, Area Under the Curve (auc) is maximum. AUC, higher the better, should be closer to 1. AUC represents, the probability of the +ve instance to be actuallly correctly predicted as +ve without any bias.

For example: If an algorithm predicts, a true +ve with AUC value 0.99, this informs that there is 99% Probability for the prediction to be correct.
