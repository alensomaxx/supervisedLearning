```python
import matplotlib.pyplot as plt##importing matplotlib library
import numpy as np##impoting numpy

# Generate data for a line plot
x = np.linspace(0, 10, 100)##synthetic data: starting from 0, ending at 10, regular intyervals
y = np.sin(x)##for each of the 100 , x values, y value

# Create a line plot
plt.plot(x, y, label='Sine wave', color='blue', linestyle='-', linewidth=2)##line plot, arguments: data declaration, aesthetics, legend

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot: Sine wave')

# Add a legend
plt.legend()##legend for the plot, it needs the label

# Show the plot
plt.show()##integrate all methods above to give the final display of the polot
```
